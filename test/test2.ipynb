```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# ====== 经典数据集 ======
from sklearn.datasets import load_iris, load_wine, load_diabetes, fetch_california_housing

iris    = load_iris()              # 鸢尾花
wine    = load_wine()              # 葡萄酒
diabetes = load_diabetes()          # 糖尿病
housing = fetch_california_housing() # 加州房价
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df_lenth_positive = iris_df.sort_values(by="sepal length (cm)")
iris_grouped = iris_df.groupby('target').mean()
iris_grouped['count'] = iris_df.groupby('target').size()

axes = iris_grouped.plot(kind='bar', subplots=True, layout=(2,3),
                         figsize=(14,10), title='Iris Dataset Grouped by Target')

for ax in axes.flat:
    if ax.containers:
        ax.bar_label(ax.containers[0], fmt='%.2f', fontsize=10)
        ymax = ax.get_ylim()[1]
        ax.set_ylim(top=ymax * 1.2)
plt.savefig('iris_grouped_bar_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
