```python
import mne
mne.set_config('MNE_DATA', '/Users/zho/code/BCI_Learning/mne_data')

from mne.datasets import eegbci
paths = eegbci.load_data(subjects=1, runs=[6, 10, 14], update_path=True)
raws = [mne.io.read_raw_edf(p, preload=True) for p in paths]
raw = mne.concatenate_raws(raws)

```

```python
print(raw.n_times)
print(raw.ch_names)
print(raw[:5, 1000:2000])
%matplotlib qt
raw.plot(duration=10, n_channels=8)
```
